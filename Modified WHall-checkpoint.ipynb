```python
import matplotlib.pyplot as plt
import math
from scipy.stats import linregress

# List of angles in degrees
twothita_degrees = [38.53, 44.17, 65.18, 78.33, 82.48, 99.12, 112.10, 116.62, 137.69]
thita_degrees = [angle/2 for angle in twothita_degrees]
thita_radians = [math.radians(angle) for angle in thita_degrees]
cosines = [math.cos(angle) for angle in thita_radians]
sines = [math.sin(angle) for angle in thita_radians]

fwhm_degrees = [0.24005, 0.29182, 0.4072, 0.5444, 0.55258, 0.80326, 0.9199, 1.07017, 1.62158]
fwhm_radians = [math.radians(angle) for angle in fwhm_degrees]

# Calculate y_bcosdata as a product of fwhm_radians and cosines
y_bcosdata = [angle * cos for angle, cos in zip(fwhm_radians, cosines)]
#y_bcos=[f"{angle*cos:.3e}" for angle,cos in zip(fwhm_radians,cosines)]---> for scientific notation

# x_sin should have the same number of elements as y_bcosdata
x_sin = sines[:len(y_bcosdata)]

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x_sin, y_bcosdata)

# Calculate the fitted line values
fit_line = [slope * x + intercept for x in x_sin]

plt.xlabel("Sin thita", fontsize=12)
plt.ylabel("B'cos thita", fontsize=12)
plt.title("Williamson-Hall plot")

# Plot the data points
plt.scatter(x_sin, y_bcosdata, label="Data", marker="o")
# Plot the fitted line
plt.plot(x_sin, fit_line, marker="o", label="Fitted Line", color='red')

plt.legend()
plt.show()

print(f"R-squared value: {r_value**2:.3f}")
print(f"Slope value: {slope:.3f}")
print(f"Intercept value: {intercept:.3f}")
print(f"Standard error i.e deviation from data: {std_err:.3e}")
```
